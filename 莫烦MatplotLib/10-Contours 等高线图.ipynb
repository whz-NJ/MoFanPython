### 
 (x,y) 256 height function f(x,y)  x, y  [-3,3] 256meshgridxy:

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
%matplotlib inline

def f(x,y):
    # the height function
    return (1 - x / 2 + x**5 + y**3) * np.exp(-x**2 -y**2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X,Y = np.meshgrid(x, y)
```

plt.contourfX, Y, f(X,Y)0.75 f(X,Y) color map

```python
# use plt.contourf to filling contours
# X, Y and value for (X,Y) point
plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap=plt.cm.hot)
```

plt.contourX, Y, f(X,Y)0.5Label

```python
# use plt.contourf to filling contours
# X, Y and value for (X,Y) point
plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap=plt.cm.hot)

# use plt.contour to add contour lines
C = plt.contour(X, Y, f(X, Y), 8, colors='black', linewidth=.5)
```
