```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
%matplotlib inline

n = 1024    # data size
X = np.random.normal(0, 1, n) # X
Y = np.random.normal(0, 1, n) # Y
T = np.arctan2(Y,X) # for color value
```

scatterplotparameter


XYlocationsize=75Tcolor mapalpha  50% x(-1.51.5)xtick()xy

```python
plt.scatter(X, Y, s=75, c=T, alpha=.5)

plt.xlim(-1.5, 1.5)
plt.xticks(())  # ignore xticks
plt.ylim(-1.5, 1.5)
plt.yticks(())  # ignore yticks

plt.show()
```
