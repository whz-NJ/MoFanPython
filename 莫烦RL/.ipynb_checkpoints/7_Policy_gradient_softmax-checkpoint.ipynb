```python
import gym
from RL_brain import PolicyGradient
import matplotlib.pyplot as plt

RENDER = False  # , 
DISPLAY_REWARD_THRESHOLD = 400  #   reward  400 

env = gym.make('CartPole-v0')   # CartPole 
env = env.unwrapped     # 
env.seed(1)     #  Policy gradient ,  variance , 

print(env.action_space)     #  action
print(env.observation_space)    #  state  observation
print(env.observation_space.high)   #  observation 
print(env.observation_space.low)    #  observation 

# 
RL = PolicyGradient(
    n_actions=env.action_space.n,
    n_features=env.observation_space.shape[0],
    learning_rate=0.02,
    reward_decay=0.99,   # gamma
    # output_graph=True,    #  tensorboard 
)
```

, ***.  Qleanring .***

```python
for i_episode in range(3000):

    observation = env.reset()

    while True:
        if RENDER: env.render()

        action = RL.choose_action(observation)

        observation_, reward, done, info = env.step(action)

        RL.store_transition(observation, action, reward)    #  transition

        if done:
            ep_rs_sum = sum(RL.ep_rs)

            if 'running_reward' not in globals():
                running_reward = ep_rs_sum
            else:
                running_reward = running_reward * 0.99 + ep_rs_sum * 0.01
            if running_reward > DISPLAY_REWARD_THRESHOLD: RENDER = True     # 
            print("episode:", i_episode, "  reward:", int(running_reward))

            vt = RL.learn() # ,  vt,  vt 

            if i_episode == 0:
                plt.plot(vt)    # plot  vt
                plt.xlabel('episode steps')
                plt.ylabel('normalized state-action value')
                plt.show()
            break

        observation = observation_
```
